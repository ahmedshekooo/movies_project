```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("./tmdb_5000_movies.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['homepage'], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.fillna({"tagline":"none"},inplace=True)
```

```python
df.isnull().sum()
```

```python
df["release_date"] = pd.to_datetime(df["release_date"])
```

```python
df.info()
```

```python
df.fillna({"release_date": "01-01-2000"}, inplace=True)
```

```python
df.isnull().sum()
```

```python
mean= df["runtime"].mean()
df.fillna({"runtime":mean}, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.fillna({"overview": "none"}, inplace=True)
```

```python
df.info()
```

```python
df.hist(figsize=(20,15), color='skyblue');
```

```python
toppop = df.nlargest(10, "popularity")
plt.style.use("dark_background")
plt.figure(figsize=(12,6))
plt.rcParams.update({'font.size': 8})
plt.bar(toppop["original_title"], toppop["popularity"], color=['skyblue', 'lightgreen', 'lightcoral', '#a39ae7'])
plt.xticks(rotation=45, ha='right')
plt.xlabel("Movie Title")
plt.ylabel("Popularity")
plt.title("Top 10 Movies by Popularity")
plt.show()
```

```python
top_languages = df["original_language"].value_counts().head(10)
plt.style.use("dark_background")
plt.figure(figsize=(12,6))
plt.bar(top_languages.index, top_languages.values, color=['skyblue', 'lightgreen', 'lightcoral', '#a39ae7'])
plt.xticks(rotation=45, ha='right')
plt.xlabel("Spoken Language")
plt.ylabel("Number of Movies")
plt.title("Top 10 Spoken Languages in Movies")
plt.show()
```

```python
import ast

# Parse genres and explode
df_genres = df.copy()
df_genres['genres'] = df_genres['genres'].apply(ast.literal_eval)
df_genres = df_genres.explode('genres')
df_genres['genre_name'] = df_genres['genres'].apply(lambda x: x['name'] if isinstance(x, dict) else None)

# Group by genre and sum revenue
genre_revenue = df_genres.groupby('genre_name')['revenue'].mean().sort_values(ascending=False).head(10)

# Plot
plt.style.use("dark_background")
plt.figure(figsize=(12,6))
plt.bar(genre_revenue.index, genre_revenue.values, color=['skyblue', 'lightgreen', 'lightcoral', '#a39ae7'])
plt.xticks(rotation=45, ha='right')
plt.xlabel("Genre")
plt.ylabel("Total Revenue")
plt.title("Top 10 Genres by Total Revenue")
plt.show()
```
